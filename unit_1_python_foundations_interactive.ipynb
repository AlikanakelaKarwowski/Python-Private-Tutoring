# Lesson 2 — Variables
## Theme: Computers Remember Things

A variable is like a labeled box.

It stores information so we can use it later.

Example:

```python
name = "Alex"
```

This means:

- make a variable called `name`,
- store the text `"Alex"` inside it.

## Example 1: Store a Name

Run this code.

````python
name = "Alex"
print(name)
````

## Example 2: Store Different Types of Information

Text uses quotation marks.

Numbers do not.

````python
character_name = "Luna"
level = 5
health = 100

print(character_name)
print(level)
print(health)
````

## Example 3: Use Variables Inside Messages

We can combine text and variables using commas inside `print()`.

````python
name = "Kai"
level = 7

print("Hero name:", name)
print("Hero level:", level)
````

Another Way to combine text and variables is by using an `fstring` or "formatted string"
You can turn any string into an `fstring` by typing "f" in front of a string like below.

`print(f"this is an fstring")`

To use variables inside of `fstring`s you wrap the variable with `{}` curly braces.

````python
name = "Riley"
level = 6
print(f"Hero name: {name}")
print(f"Hero level: {level}")
````

## Example 4: Variables Can Change

A variable can be updated.

````python
hp = 100
print("Starting HP:", hp)

hp = 70
print("After getting hit:", hp)

hp = 90
print("After healing:", hp)
````

# Lesson 4 — Math in Programming
## Theme: Computers Are Super Calculators

Python can do math.

Common math operators:

| Operator | Meaning |
|---|---|
| `+` | addition |
| `-` | subtraction |
| `*` | multiplication |
| `/` | division |

Programming math is used in:

- games,
- money apps,
- sports stats,
- science,
- animation,
- AI,
- websites.

## Example 1: Basic Math

````python
print(5 + 3)
print(10 - 4)
print(6 * 7)
print(20 / 5)
````

## Example 2: Math With Variables

````python
attack = 10
weapon_bonus = 5

damage = attack + weapon_bonus

print("Damage:", damage)
````

## Example 3: Updating Numbers

````python
gold = 100
print("Starting gold:", gold)

gold = gold + 50
print("After quest reward:", gold)

gold = gold - 30
print("After buying a sword:", gold)
````

## Example 4: Turning Input Into Numbers

`input()` gives us text.

To do math with typed numbers, use `int()`.

`int()` converts text into a whole number.

````python
age_text = input("How old are you? ")
age = int(age_text)

next_year = age + 1

print("Next year you will be", next_year)
````
